```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Housing.csv')
df.head()
df = df[df['area']<8000]
df.head()
```

```python
x = df['area']
y = df['price']
y.head()
```

```python
plt.scatter(x,y)
plt.xlabel('area')
plt.ylabel('price')
plt.title('Housing Area vs Price')
plt.show()
```

```python
x = x.to_numpy()
y = y.to_numpy()


```

```python
x_mean = x.mean()
y_mean = y.mean()
print(x_mean)
print(y_mean)
```

```python
m = ((x-x_mean)*(y-y_mean)).sum()/((x-x_mean)**2).sum()

b = y_mean - m*x_mean
print(m)
print(b)
```

```python
y_pred = m*x + b
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred,'r-')
plt.xlabel('area')
plt.ylabel('price')
plt.title('Housing Area vs Price')
plt.show()

```

```python
mse = ((y-y_pred)**2).sum()/len(y)
print(f"Mean Square Error : {mse}")
```

```python
r2_score = 1 - ((y-y_pred)**2).sum()/((y-y_mean)**2).sum()
print(f"R^2 score: {r2_score}")
```
